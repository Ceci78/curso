```python
#Ejercicio 1: Análisis de Rentabilidad de Productos

import pandas as pd

data = {
    'Producto': ['Computadora', 'Tablet', 'Smartphone', 'Auriculares', 'Cargador'],
    'Costo': [500, 300, 200, 50, 20],
    'Precio_venta': [1000, 500, 300, 100, 50],
    'Unidades_vendidas': [30, 50, 70, 100, 150]
}

df_productos = pd.DataFrame(data)

# Calcula la ganancia por producto
df_productos['Ganancia'] = (df_productos['Precio_venta'] - df_productos['Costo']) * df_productos['Unidades_vendidas']

# Calcula el costo total
df_productos['Costo_total'] = df_productos['Costo'] * df_productos['Unidades_vendidas']

# Calcula la rentabilidad como porcentaje
df_productos['Rentabilidad (%)'] = (df_productos['Ganancia'] / df_productos['Costo_total'] * 100).astype(int)

df_productos = df_productos.sort_values(by='Rentabilidad (%)', ascending=False) #Ordena los productos por rentabilidad

df_productos
```

```python
#Ejercicio 2: Segmentación de Clientes por Edad y Consumo

import pandas as pd

data_clientes = {
    'Cliente': ['Ana', 'Luis', 'María', 'Carlos', 'Josefina'],
    'Edad': [25, 35, 45, 55, 65],
    'Gasto_mensual': [200, 400, 600, 800, 1000]
}

df_clientes = pd.DataFrame(data_clientes)
df_clientes
```

```python
# Definir los rangos de edad
bins = [0, 30, 50, 100]  # Rangos de edad
labels = ['0-30', '31-50', '51-100']  # Etiquetas para los rangos

# Crear una nueva columna 'Rango de Edad'
df_clientes['Rango de Edad'] = pd.cut(df_clientes['Edad'], bins=bins, labels=labels, right=True)

df_clientes
```

```python

gasto_promedio = df_clientes['Gasto_mensual'].mean()
print(gasto_promedio)
```

```python
#Ejercicio 3: Optimización de Inventarios

data_inventario = {
    'Producto': ['Laptop', 'Mouse', 'Teclado', 'Monitor'],
    'Inventario_actual': [200, 1500, 800, 150],
    'Ventas_anuales': [600, 1800, 900, 300]
}

df_inventario = pd.DataFrame(data_inventario)
df_inventario

# Calcular la tasa de rotación de inventario
df_inventario['Tasa_rotacion'] = df_inventario['Ventas_anuales'] / df_inventario['Inventario_actual']

df_inventario['Tasa_rotacion'] = df_inventario['Tasa_rotacion'].round(1)

print(df_inventario)
```
